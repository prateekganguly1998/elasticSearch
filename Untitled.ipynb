```python
import json
import pandas as pd
```

```python
with open('yelp_academic_dataset_business.json', encoding='utf-8') as f:
    business_data= json.load(f)
```

```python
df=pd.DataFrame(business_data['businesses'])
```

```python
attributes_series=df['attributes']
print(attributes_series[])
```

```python
df['ByAppointmentOnly']=df['attributes'].apply(lambda a: cleanAttributes(a))
```

```python
def cleanAttributes(diction):
    if diction is not None:
        for key,values in diction.items():
            if key == 'ByAppointmentOnly' and values == 'True':
                return True
            elif key == 'ByAppointmentOnly' and values == 'False':
                return False        
    
```
